**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto
filtro1=df[df["region"] == "North America"]
filtro1
#SE FILTRAN LOS DATOS PARA SOLO OBSERVAR LOS PA√çSES QUE PERTENECEN A LA REGION "NORTH AMERICA"

```

```python
#Filtro por filas
filtro2=df.iloc[ 30:98 , : ]  #filas consecutivas
filtro2
#SE FILTRAN DATOS PARA QUE SE VEAN TODAS LAS COLUMNAS PERO, EN LAS FILAS, SOLO SE OBSERVEN LA INFORMACION DE LAS FILAS 30 A 97
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , [0,7,10]]  #columnas no consecutivas
filtro3
#SE FILTRAN DATOS PARA QUE SE VEAN TODAS LAS FILAS PERO, EN LAS COLUMNAS, SOLO SE OBSERVEN LOS DATOS DE LA PRIMERA, OCTAVA Y UND√âICIMA COLUMNA

```

```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
#Paso 2: Aplicamos Filtro de filas vs columnas
filtro4=df.loc[["Europe","Asia"], ["code", "name","region"]]
filtro4
#LA COLUMNA 3 DE LA TABLA (CONTINENT) SE VUELVE EL √çNDICE DE LA TABLA Y SE FILTRAN LOS DATOS QUE EN LAS FILAS SOLO SE VEA LA INFORMACION DE LOS CONTINENTES "EUROPE" Y "ASIA"
# Y EN LAS COLUMNAS SOLO SE VEA "CODE", "NAME" Y "REGION".
```

```python
#Filtro por cabecera
filtro5=df.head(25)
filtro5
#SE FILTRAN DATOS PARA SOLO APAREZCAN LAS PRIMERAS 25 FILAS

```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["life_expectancy"] > 78]
filtro6
#SE FILTRAN DATOS PARA QUE APAREZCAN LOS PA√çSES CON UN INDICE DE VIDA MAYOR A 78

```

**FILTROS BASICOS CON CONETORES LOGICOS**


```python
#Filtro Y
filtro7=df[(df["life_expectancy"] > 78) & (df["region"] == "Western Europe")]
filtro7
#SE FILTRAN DATOS PARA QUE APAREZCAN LOS PA√çSES CON UN INDICE DE VIDA MAYOR A 78 Y PA√çSES UBICADOS EN LA REGI√ìN "WESTERN EUROPE"
```

```python
#Filtro o
filtro8=df[(df["life_expectancy"] > 78) | (df["government_form"] == "Republic")]
filtro8
#SE FILTRAN DATOS PARA QUE APAREZACAN LOS PA√çSES CON UN INDICE DE VIDA MAYOR A 78 O AQUELLOS PA√çSES EN LOS QUE SU FORMA DE GOBIERNO ES REPUBLICA

```

```python
#Filtro not
filtro9=df[~(df["life_expectancy"] <= 78) & ~(df["government_form"] == "Republic")]
filtro9
#SE FILTRAN DATOS PARA QUE NO APAREZACAN LOS PA√çSES CON UN INDICE DE VIDA MENOR O IGUAL A 78 O AQUELLOS PA√çSES EN LOS QUE SU FORMA DE GOBIERNO ES REPUBLICA
```
